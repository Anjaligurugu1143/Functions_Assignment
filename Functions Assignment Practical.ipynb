## 2. Create a Python function that accepts a string and returns the reverse of that string.

```python
def str_rev(str):
    rev=""
    for i in str:
        rev=i+rev
    return rev
str_rev("python")
```

## 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

```python
def squares(l):
    sq=[]
    for i in l:
        sq.append(i**2)
    return sq
squares([2,3,4,5,6])
```

## 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def prime(num):
    if 1<=num<=200:
        count=0
        for i in range(1,num+1):
            if num%i==0:
                count=count+1
        if count==2:
            return "prime"
        return "not prime"
    else:
        print("Number is not from 1 to 200.")
prime(17)
```

## 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
def fibonacci_generator(n):
    a, b = 0, 1
    count = 0
    while count < n:
        yield a
        a, b = b, a + b
        count += 1
for number in fibonacci_generator(10):
    print(number)
```

##  6.Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_of_2(n):
    for i in range(1,n+1):
        yield 2**i
for power in power_of_2(5):
    print(power)
```

## 7.Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
for line in read_file_line_by_line('demo.txt'):
    print(line)
```

## 10.Create a Python program that uses filter() to remove all the vowels from a given string.

```python
input_string = "Hello, how are you?"
no_vowels_string = ''.join(filter(lambda char: char.lower() not in 'aeiou', input_string))
print(no_vowels_string)
```
